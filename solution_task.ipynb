# Задание №1
Напишите функцию, которая классифицирует фильмы из материалов занятия по следующим правилам:
- оценка 2 и меньше - низкий рейтинг
- оценка 4 и меньше - средний рейтинг
- оценка 4.5 и 5 - высокий рейтинг

Результат классификации запишите в столбец class

#### Первый вариант решения задачи. Клафицируем оценки фильмов от каждого юзера

```python
ratings = pd.read_csv('ratings.csv')
movies= pd.read_csv('movies.csv')
full_info_movies = ratings.merge(movies, on='movieId', how='left')
full_info_movies[['movieId', 'title', 'rating']].head()
```

```python
## Функция для классификации оценки 
def class_rating(rating):
    if rating <= 2.0:
        return 'низкий рейтинг'
    elif rating <=4.0 :
        return 'средний рейтинг'
    elif rating > 4.0:
        return 'высокий рейтинг'
```

```python
full_info_movies['class'] = full_info_movies['rating'].apply(class_rating)
full_info_movies.sort_values('rating', ascending=True)
```

#### Второй вариант решения задачи. Классифицируем среднюю оценку фильма.

```python
# Вывод средней оценки фильма 
info_avg_rate=full_info_movies[['title','rating']].groupby('title').mean()  #Подсчет среднего rating фильма
info_avg_rate['class'] = info_avg_rate['rating'].apply(class_rating)     
info_avg_rate
```

# Задание №2
Используем файл keywords.csv.

Необходимо написать гео-классификатор, который каждой строке сможет выставить географическую принадлежность определенному региону. Т. е. если поисковый запрос содержит название города региона, то в столбце ‘region’ пишется название этого региона. Если поисковый запрос не содержит названия города, то ставим ‘undefined’.

```python
import re
```

```python
keyword= pd.read_csv('keywords.csv')
```

```python
def region(keyword):
    geo_data={'Центр': ['москва', 'тула', 'ярославль'],
            'Северо-Запад': ['петербург', 'псков', 'мурманск'],
            'Дальний Восток': ['владивосток', 'сахалин', 'хабаровск']
             }
    for i in geo_data:
        for j in geo_data[i]:
            if re.findall(j[:-1] + '\w+',keyword ) != []:
                return i
    return 'undefined'
```

```python
keyword['region'] = keyword['keyword'].apply(region)
```

# Задание №3
Есть мнение, что “раньше снимали настоящее кино, не то что сейчас”. Ваша задача проверить это утверждение, используя файлы с рейтингами фильмов из прошлого домашнего занятия (файл movies.csv и ratings.csv из базы grouplens.org.../movielens). Т. е. проверить верно ли, что с ростом года выпуска фильма его средний рейтинг становится ниже.


####  Алгоритм:
1. В переменную years запишите список из всех годов с 1950 по 2010.
2. Напишите функцию production_year, которая каждой строке из названия фильма выставляет год выпуска. Не все названия фильмов содержат год выпуска в одинаковом формате, поэтому используйте следующий алгоритм:
* для каждой строки пройдите по всем годам списка years
* если номер года присутствует в названии фильма, то функция возвращает этот год как год выпуска
* если ни один из номеров года списка years не встретился в названии фильма, то возвращается 1900 год
3. Запишите год выпуска фильма по алгоритму пункта 2 в новый столбец ‘year’
4. Посчитайте средний рейтинг всех фильмов для каждого значения столбца ‘year’ и отсортируйте результат по убыванию рейтинга

```python
def year_movie(movie):
    years = [i for i in range(1950, 2011)] 
    try:
        yy = re.search(r'\d{4}',movie).group(0)
    except:
        return 1900
    if int(yy) in years:
        return yy
    else:
        return 1900
```

```python
full_info_movies['year'] = full_info_movies['title'].apply(year_movie)
```

```python
full_info_movies.sort_values('title', ascending=True)
```

```python
full_info_movies[['year','rating']].groupby('year').mean().sort_values('rating', ascending=False).head(20)
```
